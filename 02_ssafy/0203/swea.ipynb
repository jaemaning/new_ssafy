```python
# 1204 .[S/W 문제해결 기본] 1일차 - 최빈수 구하기

T = int(input())
# 여러개의 테스트 케이스가 주어지므로, 각각을 처리합니다.
for test_case in range(1, T + 1):
    test_num = int(input())
    numberList = list(map(int, input().split()))
    maxValue = max(numberList)
    countList = [0]*(maxValue+1)

    for i in numberList:
        countList[i] += 1
    
    targets=[i for i,v in enumerate(countList) if v==max(countList)]
    targetValue=max(targets)

    print(f"#{test_num} {targetValue}")
```

```python
# 1959 .두 개의 숫자열

T = int(input())
# 여러개의 테스트 케이스가 주어지므로, 각각을 처리합니다.
for test_case in range(1, T + 1):
    n,m = map(int,input().split())
    firstList = list(map(int, input().split()))
    secondList = list(map(int, input().split()))
    answer = []

    if n<=m:  # n을 이동
        move = m-n
        for i in range(move+1):
            result=0
            for j in range(n):
                result+=firstList[j]*secondList[j+i]
            answer.append(result)
            
    else:  # m을 이동
        move = n-m
        for i in range(move+1):
            result=0
            for j in range(m):
                result+=firstList[j+i]*secondList[j]
            answer.append(result)

    targetValue = max(answer)
    print(f"#{test_case} {targetValue}")
```

```python
# 1961 .숫자 배열 회전

T = int(input())
# 여러개의 테스트 케이스가 주어지므로, 각각을 처리합니다.
for test_case in range(1, T + 1):
    n = int(input())
    box = []
    for i in range(n):
        line = list(map(int,input().split()))
        box.append(line)

    # 90 도
    ans90=""
    for i in range(n):
        for j in range(n-1,-1,-1):
            ans90+=str(box[j][i])

    # 180 도
    ans180=""
    for i in range(n-1,-1,-1):
        for j in range(n-1,-1,-1):
            ans180+=str(box[i][j])

    # 270 도
    ans270=""
    for i in range(n-1,-1,-1):
        for j in range(n):
            ans270+=str(box[j][i])

    print(f"#{test_case}")
    for i in range(n):
        print(ans90[i*n:i*n+n]+ " " + ans180[i*n:i*n+n] + " " + ans270[i*n:i*n+n])
```
